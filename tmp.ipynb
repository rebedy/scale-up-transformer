```python
import torch
t = torch.tensor(
    [[2,2,3],
     [3,2,1],
     [1,3,5]],
     dtype=torch.float
     )
```

```python
indices = [list(row).index(2) if 2 in row else -1 for row in t]
for row, idx in enumerate(indices):
    if idx >= 0:
        t[row, idx+1:] = 0
```
